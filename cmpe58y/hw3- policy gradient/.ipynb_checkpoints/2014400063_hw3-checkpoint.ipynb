```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import math
from collections import deque
```

```python
class NeuralNetwork:
    def __init__(self, weights=np.random.rand(4), learning_rate = 0.05, discount_factor= 0.99):
        # Initialize we,ghts, learning rate and discount factor and environment

        self.weights = weights
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor
        self.env = gym.make('CartPole-v0')

    #sigmoid function
    def sigmoid(self, x):
        return 1.0 / (1.0 + np.exp(-x)) # sigmoid "squashing" function to interval [0,1]

    #discounted rewards for each trajectory
    def calculateDiscountedReward(self,rewards):
        discounted_r = np.zeros_like(rewards)
        running_add = 0
        for t in reversed(range(0, rewards.size)):
            running_add = running_add * self.discount_factor + rewards[t]
            discounted_r[t] = running_add
        return np.array(discounted_r)
    
    #calculating gradients, dot product of observations as states and gradients_stack as the multiplication
    # of discounted rewards and probability value of chosen actions
    def policy_backward(self,observation_stack, gradients_stack):
        grad_episode = np.dot(gradients_stack.T, observation_stack).reshape(4,1)
        return grad_episode.ravel()
    
    #update network 
    def updatedWeights(self,buffer_grad, observation_stack,gradients_stack)   :
        self.weights = self.weights + self.learning_rate*buffer_grad

    #generate action, probability of action and sub-result of gradient of policy with neural network with one neuron
    # and observation as input.
    def policy_forward(self,observation):
        result = np.dot(self.weights, observation)
        prob = self.sigmoid(result)
        probabilities =  np.array([prob, 1-prob])
        gradients = np.array([-1*prob, 1-prob])
        action = np.random.choice([0, 1], p=probabilities)
        return action,probabilities,gradients
    
    #plot episode / total reward
    def plotScores(self,scores):
        plt.plot(scores)
        plt.xlabel('episode')
        plt.ylabel('total reward')
        plt.savefig('RewardOverEpisodes.png')
        plt.show()
    
    #run one episode
    def oneEpisode(self,render):
        observation = self.env.reset()
        
        rewardOfTheEpisode = 0
        
        
        observations = []
        actions = []
        rewards = []
        probs = []
        gradients = []
        
        for i in range(500):
            if render:
                env.render()

            observations.append(observation)
            
            #get the action with respect to observation from neural network
            action, prob,gradient = self.policy_forward(observation)
            #get new observation
            observation, reward, _, _ = self.env.step(action)

            rewardOfTheEpisode += reward
            rewards.append(reward)
            actions.append(action)
            probs.append(prob)
            gradients.append(gradient[action])

        return rewardOfTheEpisode, np.array(rewards), np.array(observations), np.array(actions), np.array(probs),np.array(gradients)
        

    def runEpisodes(self,numberOfEpisodes,batch):
        
        rewardsOfEpisodes = []
        numberOfConsecutiveSolvedTasks = 0
        buffer_grad = 0
        
        for episode in range(1,numberOfEpisodes):
            
            total_reward, rewards, observations, actions, probs, gradients = self.oneEpisode(False)
            #
            rewardsOfEpisodes.append(total_reward)
            #stack together all rewards, observations, actions,probs and gradients for this episode
            rewards_stack = np.vstack(rewards)
            observation_stack = np.vstack(observations)
            actions_stack = np.vstack(actions)
            probs_stack =np.vstack(probs)
            gradients_stack = np.vstack(gradients)

            #calculate discounted reward
            discounted_reward = self.calculateDiscountedReward(rewards_stack)
            discounted_reward -= np.mean(discounted_reward)
            discounted_reward /= np.std(discounted_reward)
            #calculate prob*discounted_reward -> prob : probability of chosen action
            gradients_stack *= discounted_reward
            #gradient result of the episode
            buffer_grad += self.policy_backward(observation_stack,gradients_stack)
            #update network at the end of the batch
            if episode % batch == 0:
                self.updatedWeights(buffer_grad, observation_stack,gradients_stack)  
                buffer_grad = 0
            #print("Episode: " + str(episode) + " reward = " + str(total_reward) )
            #print(self.weights)
            #if total_reward >= 450, episode is counted as solved
            if total_reward >= 450:
                numberOfConsecutiveSolvedTasks += 1 
            else:
                numberOfConsecutiveSolvedTasks = 0
            #if number of consecutive solved episodes is bigger than 50, the task is solved
            if numberOfConsecutiveSolvedTasks >= 50:
                break
            
        self.plotScores(rewardsOfEpisodes)
        
```

```python
#initialize agent
agent = NeuralNetwork()
#run the task
agent.runEpisodes(1000,10)
```
